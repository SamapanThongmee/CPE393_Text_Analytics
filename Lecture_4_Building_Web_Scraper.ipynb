##### <b>Downloading and interoretingrobot.txt</b>

```python
from urllib.request import urlopen
res = urlopen("https://www.reuters.com/robots.txt")
print(res.read().decode('utf-8'))
```

```python
import urllib.robotparser

rp = urllib.robotparser.RobotFileParser()
rp.set_url("https://www.reuters.com/robots.txt")
rp.read()
rp.can_fetch("*", ("https://www.reuters.com/sitemap.xml"))
```

```python
rp.site_maps()
```

##### <b>Download data and libraries</b>

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.edge.options import Options
from bs4 import BeautifulSoup
from time import sleep
import re
import pandas as pd

# Your options may be different
options = Options()
# options.set_preference('permissions.default.image', 2)
# options.set_preference('dom.ipc.plugins.enabled.libflashplayer.so', False)

options.add_argument('--blink-settings=imagesEnabled=false')  # Disable images
options.add_argument('--disable-extensions')  # Disable browser extensions
```

```python
# initialize web driver

# Create the Edge WebDriver with the specified options
driver = webdriver.Edge(options=options)
```

```python
# visit target  site
url = 'https://pantip.com/forum/blueplanet'
driver.get(url)
sleep(1)
```

```python
# Select element
from bs4 import BeautifulSoup

soup = BeautifulSoup(driver.page_source, 'html.parser')

linksReviewsElem = soup.select('a.gtm-main-content-link')

# show the first 5 elements of selected element
linksReviewsElem[:5]
```

```python
len(linksReviewsElem)
```

```python
# Select element
from bs4 import BeautifulSoup

linksReviewsElem = soup.select('h2 > a.gtm-main-content-link')

# show the first 5 elements of selected element
linksReviewsElem
```

```python
# Extract attributes of selected elements

linksElem = linksReviewsElem
links = []

# iterate over element
for elem in linksElem:
    href = elem.get('href')
    links.append(href)

links
```

```python
# visit each page of topics and extract the information

titleList = []
contentList = []

for url in links:
    driver.get(url)
    sleep(0.5)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    title = soup.select('h2.display-post-title')[0]
    content = soup.select('div.display-post-story')[0]
    titleList.append(title.get_text())
    contentList.append(content.get_text())

driver.close()
data = pd.DataFrame({'url':links,
                     'title':titleList,
                     'content':contentList})

data
```
