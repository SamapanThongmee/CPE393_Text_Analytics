##### <b>Downloading and interoretingrobot.txt</b>

```python
from urllib.request import urlopen
res = urlopen("https://www.reuters.com/robots.txt")
print(res.read().decode('utf-8'))
```

```python
import urllib.robotparser

rp = urllib.robotparser.RobotFileParser()
rp.set_url("https://www.reuters.com/robots.txt")
rp.read()
rp.can_fetch("*", ("https://www.reuters.com/sitemap.xml"))
```

```python
rp.site_maps()
```

##### <b>Find URLs from sitemap.xml</b>

```python
!pip install --upgrade certifi
!pip install --upgrade requests
```

```python
import xmltodict
import requests

sitemap = xmltodict.parse(requests.get(rp.site_maps()[0], verify=False).text)
```

```python
sitemap.keys()
```

```python
import pandas as pd
sitemap_link_df = pd.DataFrame.from_dict(sitemap['sitemapindex']['sitemap'])
sitemap_link_df.head()
```
