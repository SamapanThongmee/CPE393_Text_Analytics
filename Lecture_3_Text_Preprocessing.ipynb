##### <b>Load Data</b>

```python
import nltk

nltk.download('punkt')

f = open('Lecture_3_Text_Preprocessing.txt', encoding='utf-8')
raw = f.read()
tokens = nltk.word_tokenize(raw)
```

```python
print(tokens[1000:1100])
```

##### <b>Regular Expression</b>

```python
tokens = [w.lower() for w in tokens]
```

```python
import re
# 'ed$' find last string 'ed'
[w for w in tokens if re.search('ed$', w)]
```

```python
[w for w in tokens if re.search('^..j..t..$', w)]
```

```python
[w for w in tokens if re.search('^spo.ts?$', w)]
```

```python
[w for w in tokens if re.search('^[b-f]', w)]
```

```python
[w for w in tokens if re.search('^[bdf][aue][rts]$', w)]
```

```python
[w for w in tokens if re.search('noo+$', w)]
```

```python
[w for w in tokens if re.search('noo*$', w)]
```

```python
[w for w in tokens if re.search('\.+$', w)]
```

```python
[w for w in tokens if re.search('^[0-9]{4}$', w)]
```

```python
nltk.download('treebank')
wsj = sorted(set(nltk.corpus.treebank.words()))
[w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
```

```python
[w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
```

```python
set([w for w in tokens if re.search('(ed|ing)$', w)])
```

##### <b>Regular Expression Tokenizer</b>

```python
s1 = ("Good muffins cost $3.88\nin New York. Please buy me\n"
     "two of them.\n\nThanks.")

s2 = ("Alas, it has not reained today. When, do you think, "
      "will it rain again?")

s3 = ("<p>Although this is <b>not</b> the case here, we must "
      "not relax our cigilance!</p>")
```

```python
s1
```

```python
s2
```

```python
nltk.regexp_tokenize(s2, r'[,\.\?!"]\s*', gaps=True)
```

```python
s3
```

```python
nltk.regexp_tokenize(s3, r'</?.>', gaps=False)
```

```python
nltk.regexp_tokenize(s3, r'</?.>', gaps=True)
```
