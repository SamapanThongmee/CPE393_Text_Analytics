##### <b>Load Data</b>

```python
import nltk

nltk.download('punkt')

f = open('Lecture_3_Text_Preprocessing.txt', encoding='utf-8')
raw = f.read()
tokens = nltk.word_tokenize(raw)
```

```python
print(tokens[1000:1100])
```

```python
tokens = [w.lower() for w in tokens]
```

##### <b>Regular Expression Tokenizer</b>

```python
s1 = ("Good muffins cost $3.88\nin New York. Please buy me\n"
     "two of them.\n\nThanks.")

s2 = ("Alas, it has not reained today. When, do you think, "
      "will it rain again?")

s3 = ("<p>Although this is <b>not</b> the case here, we must "
      "not relax our cigilance!</p>")
```

```python
s1
```

```python
nltk.regexp_tokenize(s1, r'[,\.\?!"]\s*', gaps=True)
```

```python
s2
```

```python
nltk.regexp_tokenize(s2, r'[,\.\?!"]\s*', gaps=True)
```

```python
s3
```

```python
nltk.regexp_tokenize(s3, r'</?.>', gaps=False)
```

```python
nltk.regexp_tokenize(s3, r'</?.>', gaps=True)
```

```python
tokenizer = nltk.RegexpTokenizer(r"\w+")
tokens_0 = tokenizer.tokenize(raw)
print(tokens_0[1000:1100])
```

```python
# case normalization

tokens = [w.lower() for w in tokens]
tokens
```

```python
# Replacing \xad

set([w.replace('\xad','') for w in tokens if w.startswith('gry')])
token_1 = [w.replace('\xad','') for w in tokens]
print(token_1[1000:1100])
```

```python
# Stemming

porter = nltk.PorterStemmer()
token_2 = [porter.stem(t) for t in token_1]
print(token_2[1000:1100])
```

```python
# Lemmatization

# wordnnet lemmatizer

import nltk
nltk.download('omw-1.4')
nltk.download('wordnet')
```

```python
wnl = nltk.WordNetLemmatizer()

print('rocks', wnl.lemmatize('rocks'))
print('corpora', wnl.lemmatize('co'))
print('running', wnl.lemmatize('running', pos='v'))
```

```python
# Listing stopwords
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print('Stopwords', stop_words)
print('Numer of stopwords: ', len(stop_words))
```

```python
token_4 = [w for w in tokens if not w in stop_words]
print(tokens[1000:1100])
print(token_4[1000:1100])
```

```python
# Part of Speech (POS) tagging
nltk.download('averaged_perceptron_tagger')
```

```python
tokens_5 = nltk.pos_tag(tokens)
```
