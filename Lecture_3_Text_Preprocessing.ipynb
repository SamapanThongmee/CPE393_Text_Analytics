##### <b>Load Data</b>

```python
import nltk

nltk.download('punkt')

f = open('Lecture_3_Text_Preprocessing.txt', encoding='utf-8')
raw = f.read()
tokens = nltk.word_tokenize(raw)
```

```python
print(tokens[1000:1100])
```

##### <b>Regular Expression</b>

```python
tokens = [w.lower() for w in tokens]
```

```python
import re
# 'ed$' find last string 'ed'
[w for w in tokens if re.search('ed$', w)]
```

```python
[w for w in tokens if re.search('^..j..t..$', w)]
```

```python
[w for w in tokens if re.search('^spo.ts?$', w)]
```

```python
[w for w in tokens if re.search('^[b-f]', w)]
```

```python
[w for w in tokens if re.search('^[bdf][aue][rts]$', w)]
```

```python
[w for w in tokens if re.search('noo+$', w)]
```

```python
[w for w in tokens if re.search('noo*$', w)]
```
